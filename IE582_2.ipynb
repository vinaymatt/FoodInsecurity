```python
X_var=df.drop(columns=["Food Insecurity Rate", "Depression Prevalence (%)",	"Chronic Kidney Disease Prevalence (%)", "Asthma Prevalence (%)","Atrial Fibrillation Prevalence (%)","Cancer Prevalence (%)","Osteoporosis Prevalence (%)"
,"Alcohol Abuse Prevalence (%)","Weighted Annual Food Budget Shortfall","Hypertension Prevalence (%)","Heart Failure Prevalence (%)","HIV/AIDS Prevalence (%)"
,"Hepatitis (Chronic Viral B & C) Prevalence (%)","Diabetes Diabetes Prevalence (%)","Alzheimer's Disease/Dementia Prevalence (%)",'bias', 'state_code'])
y_var=df[["Food Insecurity Rate"]]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from sklearn import set_config
```

```python
X_var.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_var, y_var, test_size = 0.3, random_state = 0)
```

```python
rfr = RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=500, max_features='auto', max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.0,
                       min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=1000, n_jobs=None, oob_score=False,
                      random_state=None, verbose=0, warm_start=False)
rfr.fit(X_train, y_train)

score = rfr.score(X_train, y_train)
print("R-squared:", score)
```

```python
#0.95 without state/ with state 0.96
```

```python

ypred = rfr.predict(X_test)
mse = mean_squared_error(y_test, ypred)
print("MSE: ", mse)
print("RMSE: ", mse*(1/2.0))
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, ypred, label="predicted")
plt.title(" test and predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```
